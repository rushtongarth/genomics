```python
import os,os.path as osp,glob
import numpy as np
import pandas as pd
```

```python
DATA = osp.abspath(osp.join(os.pardir,'data'))
CHRM = glob.glob(osp.join(DATA,'chrom-*'))
CH_D = {osp.split(k)[-1]:os.listdir(k) for k in CHRM}
CH_D['chrom-01'][0],len(CH_D['chrom-01'])
```

```python
def res_path(k,v):
  return osp.join(DATA,k,v)
def get_colvals(k,v,col):
  df = pd.read_pickle(res_path(k,v))
  return getattr(df,col).values
```
